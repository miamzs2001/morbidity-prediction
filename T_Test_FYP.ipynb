```python
import numpy as np
from scipy import stats
import pandas as pd
rng = np.random.default_rng()
```

```python
df = pd.read_csv("Outcome_2.csv")
```

```python
#Select non-binary columns
mask = (df != 0) & (df != 1) & df.notnull()
continuous = df.loc[:, mask.any()]
```

```python
continuous = pd.concat((continuous,df['Outcome']),axis=1)
continuous
```

```python
#t-test with scipy
from scipy import stats

results = {}
for col in continuous.columns[:-1]:
    tstat, pvalue = stats.ttest_ind(continuous[col], continuous.iloc[:, -1])
    results[col] = (tstat, pvalue)

results = pd.DataFrame.from_dict(results, orient='index', columns=['t-statistic', 'p-value'])
results
```
